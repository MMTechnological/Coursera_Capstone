### 3. Scraping a specific webpage

The goal is to obtain the data that is in the table of postal codes and to transform the data into a pandas dataframe

The dataframe will consist of three columns: Postal Code, Borough, and Neighborhood

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
# print(table.prettify())
print('soup ready')

from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

### 4. Convert HTML table to a Pandas Dataframe to clean and preprocess

```python
dfs = pd.read_html(tab)
df=dfs[0]
df.head()
```

### 5. Data preprocessing and cleaning

```python
# Dropping the rows where Borough is 'Not assigned'
df1 = df[df.Borough != 'Not assigned']

# Combining the neighbourhoods with same Postalcode
df2 = df1.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

# Replacing the name of the neighbourhoods which are 'Not assigned' with names of Borough
df2['Neighborhood'] = np.where(df2['Neighborhood'] == 'Not assigned',df2['Borough'], df2['Neighborhood'])

df2
```

### 7. Import .CSV file with geographical data of Canada's neighbourhoods

```python
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

### 8. Merging tables (Postal code, Boroughs and Neighbourhood + geographical locations)

```python
df3 = pd.merge(df2,lat_lon)
df3.head(12)
```

### 9. Exploring and cluster the neighborhoods in Toronto

```python
# Getting all the rows that have Toronto on the Borough
df4 = df3[df3['Borough'].str.contains('Toronto',regex=False)]
df4
```

### 9.1. Visualize the previous dataframe using Folium

```python
#import folium for visualizations and KMeans for clustering neighbourhoods

#!conda install -c conda-forge folium=0.5.0 --yes (too slow, had to try installing in a different way)
!pip install folium #plotting library installed quickly
import folium # plotting library (useless line of code)

from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors

print('folium imported')
```

```python
#Build Toronto's borough's map

map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighborhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

### 9.2. Clustering the neighbourhoods in Toronto

```python
#use KMeans to cluster
k=5
toronto_clustering = df4.drop(['Postal Code','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df4.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighborhood'], df4['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
